## Por√≥wnanie SQL z Pandas

```python
# SELECT * FROM retail;
# wydobƒÖd≈∫ wszystkie dane z obiektu DataFrame, w Pandas to tylko wywo≈Çanie obiektu

df
```

```python
# SELECT Quantity, UnitPrice, CustomerID FROM retial;
# odpytanie tabeli o poszczeg√≥lne kolumny
# przekazujemy df i podajemy w [[]] listƒô kolumn kt√≥re chcƒô wyciƒÖƒá

df[['Quantity', 'UnitPrice', 'CustomerID']]

```

```python
# SELECT Quantity, UnitPrice, CustomerID FROM retial LIMIT 10;
# odpytanie tabeli o poszczeg√≥lne kolumny
# przekazujemy df i podajemy w [[]] listƒô kolumn kt√≥re chcƒô wyciƒÖƒá

df[['Quantity', 'UnitPrice', 'CustomerID']].head(10)
```

```python
# lub

df[['Quantity', 'UnitPrice', 'CustomerID']][:10]
```

```python
# SELECT * FROM retial WHERE CustomerID='17850';

df[df.CustomerID == '17850']
```

```python
# lub query

df.query('CustomerID == "17850"')
```

```python
# instrukcje warunkowe
# SELECT * FROM retial WHERE CustomerID='17850' and UnitPrice > 5;
# dwie maski logiczne w oddzielnych nawiasach

df[(df.CustomerID == '17850') & (df.UnitPrice > 5)]
```

```python
# SELECT * FROM retial WHERE CustomerID='17850' or Country='France';

df[(df.CustomerID == '17850') | (df.Country == 'France')]
```

```python
# SELECT * FROM retail WHERE InvoiceNo is not null;

df[df.InvoiceNo.notnull()]
```

```python
# isnull jak chcƒô wyciƒÖƒá braki danych

df[df.InvoiceNo.isnull()]

#. nie ma brak√≥w danych w tej kolumnie
```
